```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import re
from io import StringIO
import os
from os.path import basename, splitext
#from pathlib import Path
import seaborn as sns
import numpy as np
```

```python
#fonction de récupération des matrices des fichiers .md
def get_sections(s):
    for sec in s.split('## '):
        yield sec if sec.startswith('## ') else '## '+sec
```

```python
#fonction de production des donnnées sous forme de tableau propre
def prod_DF(texteT,titre):
    dfGlobalDataLoc = pd.DataFrame(columns=['accuracy', 'precision', 'recall','support','label'])
    for i,sec in enumerate(get_sections(texteT)):
        if i == 0:
            pass
        else:
            print(i,sec)
            sec2 = re.sub('##.*', '', sec)
            currentVal = re.search('##\s*(\w*)\n', sec).group(1)
            df = pd.read_csv(StringIO(sec2.replace(' ', '')), sep='|',index_col=1).dropna(axis=1,how='all').iloc[1:]
            df.insert(4,"deaf",titre,True)
            df.insert(5,"label",currentVal,True)
            dfGlobalDataLoc = dfGlobalDataLoc.append(df)
    print(dfGlobalDataLoc)        
    return dfGlobalDataLoc
```

```python
## source : https://www.alliage-ad.com/data-science/comment-etudier-la-liaison-entre-une-variable-quantitative-et-une-variable-qualitative/
def rapport_corr(x,y):
    somme = 0
    for facteur in x.value_counts().index:
        ind = x[x == facteur].dropna(axis=0).index
        classe = y.iloc[ind,]
        somme = somme + len(classe)*np.sum((np.mean(classe)-np.mean(y))**2)
        y_ecart = np.sum((y-np.mean(y))**2)
        rc = somme/y_ecart
        #rc.index = ['Rapport de Correlation']
        return rc
```

```python
##test Anova et kruskal : même source
import scipy.stats as stats
```

```python
def prod_DFT(texteT,titre,subfile):
    dfGlobalDataLoc = pd.DataFrame(columns=['accuracy', 'precision', 'recall','support','typology','label'])
    for i,sec in enumerate(get_sections(texteT)):
        if i == 0:
            pass
        else:
            print(i,sec)
            sec2 = re.sub('##.*', '', sec)
            currentVal = re.search('##\s*(\w*)\n', sec).group(1)
            df = pd.read_csv(StringIO(sec2.replace(' ', '')), sep='|',index_col=1).dropna(axis=1,how='all').iloc[1:]
            df.insert(4,"type",subfile, True)
            titreOk =  titre.partition(".tsv")[0]
            #variation tidyLex
            #titreOk =  titre.partition("TidyLex")[0]
            df.insert(5,"typology",titreOk,True)
            df.insert(6,"label",currentVal,True)
            dfGlobalDataLoc = dfGlobalDataLoc.append(df)
    print(dfGlobalDataLoc)        
    return dfGlobalDataLoc
```

```python
dfGlobalDataT = pd.DataFrame(columns=['accuracy', 'precision', 'recall','support','type','typology','label'])
basepath = 'tableursParVal/'
for subFile in os.listdir(basepath):
    if os.path.isdir(os.path.join(basepath, subFile)):
        ## attention aux noms : variations TidyLex
        dir_path = os.path.join(basepath, subFile)+'/'
        print(dir_path)
        for entry in os.listdir(dir_path):
            if entry.endswith('Confusion.md'):
                fullPath =dir_path+entry
                with open(fullPath, 'r') as f:
                    texteT = f.read()
                    print(f)
                    t = splitext(basename(fullPath))[0]
                    titre = t.partition(".tsv")[0]
                    dfGlobalDataLoc = prod_DFT(texteT,titre,subFile)
                    dfGlobalDataT = dfGlobalDataT.append(dfGlobalDataLoc)
```

```python
dfGlobalDataT = dfGlobalDataT.loc["all"]
dfGlobalDataT
```

```python
Dcompo = dfGlobalDataT[dfGlobalDataT['type'] == 'dateCompositionType']
Dms = dfGlobalDataT[dfGlobalDataT['type'] == 'dateManuscritType']
Lcompo = dfGlobalDataT[dfGlobalDataT['type'] == 'lieuCompositionType']
Lms = dfGlobalDataT[dfGlobalDataT['type'] == 'lieuManuscritType']
genre = dfGlobalDataT[dfGlobalDataT['type'] == 'genreType']
Dcompo
```

```python
anova_t,anova_pvalue = stats.f_oneway(Dcompo.accuracy,Dms.accuracy,Lcompo.accuracy,Lms.accuracy,genre.accuracy)
kruskal_t,kruskal_pvalue = stats.kruskal(Dcompo.accuracy,Dms.accuracy,Lcompo.accuracy,Lms.accuracy,genre.accuracy)
print(f'Anova test : statistique = {anova_t} ; pvalue = {anova_pvalue} \nKruskal-Wallis test : statistique = {kruskal_t} ; pvalue = {kruskal_pvalue}')
```
