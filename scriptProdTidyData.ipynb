```python
import pandas as pd
import os
import re
from os.path import basename, splitext
import subprocess
import errno
```

```python
def productionFileTidy(tableau,subFile,titre):
    #dataLex = tableau.loc[(tableau.lemma != "Ref.") & (tableau.POS != "PONfbl") & (tableau.POS != "PONfrt")]
    ##tidyLex
    dataLex = tableau.loc[(tableau.lemma != "Ref.") & (tableau.POS != "PONfbl") & (tableau.POS != "PONfrt") & ((tableau.POS == "NOMcom") | (tableau.POS == "VERcjg") | (tableau.POS == "VERinf") | (tableau.POS == "VERppe") | (tableau.POS == "VERppa") | (tableau.POS == "ADJqua"))] 
    try: 
        os.makedirs('tableursParVal/'+subFile+'/tidyLex/')
    except OSError as exc: 
        if exc.errno != errno.EEXIST: 
            raise 
        pass
    dataLex.to_csv('tableursParVal/'+subFile+'/tidyLex/'+titre+'TidyLex.tsv',header=True,sep='\t', index=False, mode='w', encoding='UTF-8')
    
```

```python
##test
with open('tableursParVal/genreType/roman.tsv', 'r') as f:
    df = pd.read_csv(f, sep='\t')
    t = splitext(basename('tableursParVal/genreType/roman.tsv'))[0]
    titre = t.partition(".tsv")[0]
    subFile = 'genreType'
    productionFileTidy(df,subFile,titre)
```

```python
basepath = 'tableursParVal/'
for subFile in os.listdir(basepath):
    if os.path.isdir(os.path.join(basepath, subFile)):
        dir_path = os.path.join(basepath, subFile)+'/'
        print(dir_path)
        for entry in os.listdir(dir_path):
            if entry.endswith('.tsv') and not entry.endswith('MorphCat.tsv'):
                fullPath =dir_path+entry
                with open(fullPath, 'r') as f:
                    df = pd.read_csv(f, sep='\t')
                    t = splitext(basename(fullPath))[0]
                    titre = t.partition("TidyLex.tsv")[0]
                    productionFileTidy(df,subFile,titre)
                    print(df)
```
